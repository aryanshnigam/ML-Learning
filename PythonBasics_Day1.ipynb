### Input & Output


```python
n=int(input("Enter a number: "))
print("The number you entered is:", n)
```

###  Conditions (if-elif-else)


```python
age = int(input("Enter your age: "))
if age >= 18:
    print("Adult")
elif age > 13:
    print("Teenager")
else:
    print("Child")
```

### Loops 


```python
for i in range(5):
    print("Iteration:", i, end = ' ')

i = 1
while i <= 5:
    print("\ncount :", i, end = ' ')
    i +=1
```

### Built in data structures (Lists, Sets, Tuples, Dictionaries)

#### Lists


```python
fruits = ["apple", "banana", "mango"]
fruits.append("grapes")
print(fruits)
print(fruits[1])
```

#### Tuples

```python
my_tuple = (10, 20, 30)
print(my_tuple)
```

#### Sets

```python
nums = {1, 2, 3, 3, 2}
print(nums)  # shows unique values
```

#### Dictionaries

```python
marks = {"Harry": 85, "Sam": 90, "Lily": 78}
print(marks["Sam"])
marks["Harry"] = 88
print(marks)
 
```

## Practice Questions

### Problem 1: Find the largest of the three numbers.

```python
a, b, c = 10, 20, 5
largest = max(a, b, c)
print("Largest:", largest)
```



### Problem 2: Check even/odd


```python
num = int(input("Enter number: "))
if num % 2 == 0:
    print("Even")
else:
    print("Odd")
```

### Problem 3: Sum of elements in a list


```python
arr = [2, 5, 7, 1]
total = sum(arr)
print(total)
```

### Problem 4: Count positives and negatives in a list


```python
arr = [-3, 4, -1, 6, 0]
pos = neg = 0
for x in arr:
    if x > 0:
        pos += 1
    elif x < 0:
        neg += 1
print("Positives:", pos, "| Negatives:", neg)
```

### Problem 5: Print a simple pattern


```python
for i in range (1,6):
    print("*"*i)
```
